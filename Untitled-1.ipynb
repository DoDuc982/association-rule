```python
import mlxtend
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
import pandas as pd
```

```python
# Đọc dữ liệu từ file excel.
# Dữ liệu lấy từ bộ dữ liệu http://archive.ics.uci.edu/ml/dfsets/Online+Retail

df = pd.read_excel("D:\PYTHON\Association rule\Online Retail.xlsx")
# Hiển thị 10 cột dữ liệu
pd.set_option('display.max_columns', 10)
df
```

```python
# Lấy ra tất cả cột của dữ liệu. Các cột vao gồm
# ['InvoiceNo', 'StockCode', 'Description', 'Quantity', 'InvoiceDate','UnitPrice', 'CustomerID', 'Country'],
# Hóa đơn số - Mã cổ phẩn - Mô tả - Số lượng - Ngày lập hóa đơn - Giá mỗi sản phẩm - Mã khách hàng - Nước
df.columns
```

```python
# Thực hiện tiền xử lý dữ liệu

# loại bỏ các khoảng trống ở đầu và cuối ở cột mô tả
df['Description'] = df['Description'].str.strip()
# Loại bỏ tất cả các ô không có mã hóa đơn, inplace = True thay đổi trực tiếp trên df
df.dropna(axis=0, subset=['InvoiceNo'], inplace=True)
df['InvoiceNo'] = df['InvoiceNo'].astype('str')
# Loại bỏ các giao dịch bắt đầu bằng chữ c (cancel)
df = df[~df['InvoiceNo'].str.contains('C')]
```

```python
# Lấy ra tất cả các nước từ dữ liệu
# Danh sách các nước bao gồm
# 'United Kingdom' 'France' 'Australia' 'Netherlands' 'Germany' 'Norway',...
country = df.Country.unique()
country
```

```python
#Chọn một số trong các nước trên đánh thứ tự từ 0-37
x = 5
basket = (df[df["Country"] == country[x]]
          .groupby(["InvoiceNo", "Description"])["Quantity"]
          .sum().unstack().fillna(0))
basket
```

```python
def check(x):
    if(x <= 0):
        return False
    else:
        return True
```

```python
basket_check = basket.applymap(check)
basket = basket_check
basket.head()
```

```python
frq_items = apriori(basket, min_support=0.15, use_colnames=True)
pd.set_option('display.max_colwidth', 1000)
frq_items
# Collecting the inferred rules in a dataframe
```

```python
rules = association_rules(frq_items, metric="confidence", min_threshold=0.6)
rules
```
